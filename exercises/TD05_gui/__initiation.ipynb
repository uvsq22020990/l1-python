```python
"""question 1 et 2"""

from tkinter import * #import tout ce qu"il y a dans x (ici x: tkinter)

CANVAS_WIDTH, CANVAS_HEIGHT = 600, 400

if __name__ == '__main__': 
    root = Tk()  #création de la fonction racine

    canvas = Canvas(root, width = CANVAS_WIDTH, height = CANVAS_HEIGHT)  #création du tableau avec dimension 
    #pourquoi pas tk.canvas --> car import * et non import tk (on import all)

    # Début de votre code
    x0 = 100
    x1 = CANVAS_WIDTH - 100
    y = CANVAS_HEIGHT / 2
    canvas.create_line(x0, y, x1, y)  #commence aux coordonnées (100,200) finit aux coord (500,200)
    canvas.create_oval(x0 - 50, y + 50, x0 + 50, y - 50) #dessine un oval qui a pour coordonnées (x1,y1) le coin en haut à gauche
    #d'un recttangle, et (x2,y2= le coin en bas à droite d'un rectange)
    canvas.create_oval(x1 - 50, y + 50, x1 + 50, y - 50)
    canvas.create_oval((x0 + x1) / 2 - 50, y + 50, (x0 + x1) / 2 + 50, y - 50)
    
    # Fin de votre code

    canvas.pack() #gestionnaire de position (en fonction des coordonnées)
    root.mainloop()  # Lancement de la boucle principale
```

```python
"""question 3"""

from tkinter import * #pourquoi pas import tk

CANVAS_WIDTH, CANVAS_HEIGHT = 600, 400


root = Tk()  #création de la fonction racine

canvas = Canvas(root, width = CANVAS_WIDTH, height = CANVAS_HEIGHT)  #création du tableau avec dimension 


# Début de votre code
x0 = 100
x1 = CANVAS_WIDTH - 100
y = CANVAS_HEIGHT / 2
canvas.create_line(x0, y, x1, y)  #commence aux coordonnées (100,200) finit aux coord (500,200)
canvas.create_oval(x0 - 50, y + 50, x0 + 50, y - 50)
canvas.create_oval(x1 - 50, y + 50, x1 + 50, y - 50)
canvas.create_oval((x0 + x1) / 2 - 50, y + 50, (x0 + x1) / 2 + 50, y - 50)
    
# Fin de votre code

canvas.grid() #Ce gestionnaire place le widget dans une case du widget parent qui est divisé en lignes et colonnes.
#ici grid marche car on fait des dessins dans un widget : canvas qu'on rajoute ensuite à la racine (root)
root.mainloop()  # Lancement de la boucle principale
```
